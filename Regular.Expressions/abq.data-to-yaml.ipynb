# Converting data from ABQ website to YAML

[![Binder](https://mybinder.org/badge_logo.svg)](https://mybinder.org/v2/gh/rwcitek/MyBinder.demo/main?labpath=%2FRegular.Expressions%2Fabq.data-to-yaml.ipynb)

## Get the data

```bash
cd /tmp/
wget http://data.cabq.gov/airquality/aqindex/history/042222.0017
ls -la 0422*
```

```bash
head -40 042222.0017 | cat -n
```

```bash
cat 042222.0017 |
sed -E '/^[A-Z]{3}/s/,/: /' |                  # replaces the ',' with a ':' for those lines that contain
sed -E 's/ *,/,/' |                            #  
sed -E '/^BEGIN_DATA/,/END_DATA/s/^/- /' |
sed -E 's/^- (BEGIN_DATA)/DATA:/' |
sed -E '/^- END_DATA/d' |
sed -E '1,10s/^(BEGIN_GROUP)/GROUPS:\n\1/' |   #
sed -E '/^BEGIN_GROUP/,/END_GROUP/s/^/  /' |
sed -E 's/^  (BEGIN_GROUP)/- \1:/' |
sed -E '/^  END_GROUP/d' |
sed -E '/^BEGIN_FILE/,/END_FILE/s/^/  /' |
sed -E 's/^  (BEGIN_FILE)/- \1:/' |
sed -E '/^  END_FILE/d' > 042222.0017.yaml
```

```bash
head -40 042222.0017.yaml
```

```bash
cat 042222.0017.yaml |
ruby -rjson -ryaml -e 'puts YAML::load(ARGF.read).to_json' |
jq . |
tee 042222.0017.json |
grep -v null, |
ruby -rjson -ryaml -e 'puts JSON.parse(ARGF.read).to_yaml(line_width: -1)' |
head -40
```

```bash
jq '.[0].GROUPS[].VARIABLE' 042222.0017.json
```
